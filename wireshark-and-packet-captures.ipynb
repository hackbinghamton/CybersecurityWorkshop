## Network Packets
### What are network packets?

When you send an email, a text message, or any other file, what does it look like traveling across a network? 
Data travelling over a network is broken down into *network packets* which are reassembled to the original data chunk once they reach their destination.

### What's in a packet?

The structure of a packet depends on the type of packet it is (file download, website information, network ping, etc.) and on the network protocol that it uses. Normally, a packet has a header and a payload.

The header keeps overhead information about the packet. For example, data transfering over the Internet requires the data to be broken down into IP packets. These data packets include:

* The source IP address
* The destination IP address
* The sequence number
* The type of service
* Flags
* Other technical data and options

Since the header is overhead data, the payload is the data being carried across the network and accounts for the bulk of the packet.

### What are packet captures?

A network packet capture, or *packet capture* for short, is a snapshot of network packets crossing a specific point in a data network.
A packet capture's scope ranges from a portion of a data packet to long lists of traffic flowing through a network from multiple sources.

Once a packet capture is taken in real-time, the user can view the traffic and analyze it.
Packets are captured and examined to help diagnose and solve network problems such as identifying security threats, network congestion, and data/packet loss.

We can use tools like Wireshark to look deeper into these!

## Wireshark

### What is *Wireshark*?

[Wireshark](https://www.wireshark.org) is open-source network analysis software used by cyber security and network professionals to take investigate network traffic!

### How to use Wireshark

#### Installing Wireshark

To install Wireshark:

* **Linux** - See if there's a package for your distribution (e.g. `sudo apt install wireshark`)
* **Windows/MacOS** - Go to the link above and follow the directions 

#### Capturing Packets

After downloading and installing Wireshark, you can launch it and double-click the name of a network interface under *Capture* to start capturing packets on that interface.

For example, click on your WiFi option to begin capturing packets travelling over your WiFi in real-time. You can also filter to capture specific packets using filter options we'll cover later.

*Note:* **Do *not* capture others' traffic in places where you don't have explicit permission.** Disable promiscuous mode to only capture packets related to your computer -- this is okay regardless of where you are. Also, even though promiscuous mode is enabled by default, if your WiFi uses WPA2 for encryption, you won't be able to see others' traffic.

image.png

You'll start to see packets highlighted in a variety of different colors, since Wireshark uses colors to help you identify the different types of traffic. By default, light purple is TCP traffic, light blue is UDP traffic, and black identifies packets with errors, like if they were delivered out-of-order.

If you'd like to learn more about the acronyms we've just used, check out our [Computer Networks Workshop](https://github.com/HackBinghamton/ComputerNetworksWorkshop)!

#### Filtering Packets

If youre trying to inspect something specific, such as the traffic a program sends, you might try to close down all other applications using the network so you can narrow down the traffic. Still, youll likely have a large amount of packets to sift through. Thats where Wiresharks filters come in.

You can apply a filter by typing it into the filter box at the top of the window and clicking Apply (or pressing Enter). For example, type dns and youll see only DNS packets. When you start typing, Wireshark will help you autocomplete your filter.

image.png

You can also click Analyze > Display Filters to choose a filter from among the default filters included in Wireshark. From here, you can add your own custom filters and save them to easily access them in the future.

image.png

#### Follow TCP Stream

Another interesting thing you can do is right-click on a packet and select Follow > TCP Stream.

Youll see the full TCP conversation between the client and the server. You can also click other protocols in the Follow menu to see the full conversations for other protocols, if applicable.

This is super handy for looking through HTTP conversations that involve multiple transactions (e.g. a website with tons of images and linked files).
